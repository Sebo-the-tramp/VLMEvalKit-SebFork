```python
import json
import pandas as pd

df_all_vlms = pd.read_csv('table.csv', sep=';')

df_all_vlms_results = pd.read_json('OpenVLM.json')
new_columns = [data for data in df_all_vlms_results.iloc[0,1:]['results'] if "Overall" in df_all_vlms_results.iloc[0,1:]['results'][data]]
df_all_vlms[new_columns] = None

with open('OpenVLM.json', 'r') as f:
    all_vlms_results = json.load(f)['results']
    for model_name in all_vlms_results:                
        for key, value in all_vlms_results[model_name].items():
            if("Overall" in value):                
                df_all_vlms.loc[df_all_vlms['LLM'] == model_name, key] = value['Overall']

df_all_vlms
```
